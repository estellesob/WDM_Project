```python
# load ontology td4
onto_path = "../data/exercises/family.owl"
onto = get_ontology(onto_path).load()
```

```python
with onto:
    # class and rule definition
    class oldPerson(Thing):
        pass

    # SWRL rule : Person > 60yo
    rule = Imp()
    rule.set_as_rule("Person(?p), age(?p, ?a), greaterThan(?a, 60) -> oldPerson(?p)")

    # Pre-reasoning cleaning (otherwise errors)
    # Rename temporarily individuals with accents to avoid Java errors
    for i in onto.individuals():
        if "é" in i.name:
            i.name = i.name.replace("é", "e")
        if "ô" in i.name:
            i.name = i.name.replace("ô", "o")
```

```python
# Visu before reasoning
print(f"\nIndividuals 'oldPerson' before reasoning : {onto.oldPerson.instances()}")
```

```python
# Reasoning
try:
    sync_reasoner_pellet(infer_property_values=True, infer_data_property_values=True)   
    print("Reasoning done.")
except Exception as e:
    print(f"Failed, Error : {e}")
```

```python
print("INFERENCE RESULTS:")

# Get all individuals that were inferred to be in the 'oldPerson' class
old_people = onto.oldPerson.instances()

if not old_people:
    print("No individuals found. Check if the SWRL rule was applied correctly.")
else:
    print(f"Found {len(old_people)} person(s) over 60 years old:")
    for person in old_people:
        # Get the age value (handling both functional and non-functional properties)
        raw_age = person.age
        
        # Logic to extract the integer value
        if isinstance(raw_age, list):
            age_val = raw_age[0] if len(raw_age) > 0 else "Unknown"
        else:
            age_val = raw_age if raw_age is not None else "Unknown"
            
        print(f"{person.name} is now classified as an 'oldPerson' (Age: {age_val})")
```
